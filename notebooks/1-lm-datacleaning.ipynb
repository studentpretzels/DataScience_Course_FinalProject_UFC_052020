Cleanup the data

```
df.isnull().values.sum()
```

```
categorical_columns_missing = ['Referee']
impute_missing=SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
df[categorical_columns_missing] = impute_missing.fit_transform(df[categorical_columns_missing])
```

```
numerical_columns_missing = ['R_Height_cms', 'R_Reach_cms', 'R_Weight_lbs','B_age','R_age'] #these are the columns that are (somewhat) important for further analysis and therefore have to be cleaned up. There are, as shown in the cell above, several columns with NaN-values but they are not of importance for further analysis or can be 0 by purpose (like e.g. B_draw)
impute_missing=SimpleImputer(missing_values=np.NaN, strategy='mean')
df[numerical_columns_missing] = impute_missing.fit_transform(df[numerical_columns_missing])
```

```
df.to_csv('data_clean.csv')
files.download("data_clean.csv")
df = pd.read_csv('data_clean.csv')
```
